```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = r'C:\Users\kiabi\Downloads\INFLUD-30-06-2020.csv'
df = pd.read_csv(filename, delimiter=';', 
                 encoding='ISO-8859-1',
                 error_bad_lines=False)
```

```python
df.groupby('SG_UF_NOT').size().sort_values().tail(5).plot(kind='bar', color='blue',figsize=(7,7), fontsize=(13))
plt.title('Estados com mais casos de Covid-19 no Brasil', fontsize=(13))
plt.xlabel('Casos', fontsize=(13))
plt.ylabel('Estados', fontsize=(13))
plt.xticks(rotation=0)
```

```python
df.SG_UF_NOT.value_counts().head(5).to_frame('Estados')
```

```python
df.groupby('ID_MUNICIP').size().sort_values().tail(5).plot(kind='barh', color='blue',figsize=(7,7), fontsize=(13))
plt.title('Municípios com mais casos de Covid-19', fontsize=(13))
plt.xlabel('Casos', fontsize=(13))
plt.ylabel('Municípios', fontsize=(13))
```

```python
df.ID_MUNICIP.value_counts().head(5).to_frame('Municípios')
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('SEM_NOT').size().plot(kind='bar',color='blue',figsize=(10,10), fontsize=(13))
plt.title('Casos de Covid-19 por semana em São José dos Campos', fontsize=(13))
plt.xlabel('Semana', fontsize=(13))
plt.ylabel('Casos', fontsize=(13))
plt.xticks(rotation=0)
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('SEM_NOT').size().sort_values()
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('ID_UNIDADE').size().sort_values().tail(5).plot(kind='barh', color='BLUE',figsize=(9,9), fontsize=(13))
plt.title('Casos nos Hospitais de São José dos Campos', fontsize=(13))
plt.xlabel('Casos', fontsize=(13))
plt.ylabel('Hospitais', fontsize=(13))
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('ID_UNIDADE').size().sort_values().tail(5)
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('CS_SEXO').size().sort_values().plot(kind='bar', color='BLUE', fontsize=(13))
plt.title('Casos por Sexo em São José dos Campos', fontsize=(13))
plt.ylabel('Casos', fontsize=(13))
plt.xlabel('Sexo', fontsize=(13))
plt.xticks(rotation=0)
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('CS_SEXO').size().sort_values().tail(5)
```

```python
df.query('ID_MUNICIP == "SAO JOSE DOS CAMPOS"').groupby('NU_IDADE_N').size().sort_values().tail(5).plot(kind='bar', color='BLUE', fontsize=(13))
plt.title('As 5 idades com mais casos em São José dos Campos', fontsize=(13))
plt.ylabel('Casos', fontsize=(13))
plt.xlabel('Idade', fontsize=(13))
plt.xticks(rotation=0)
```
